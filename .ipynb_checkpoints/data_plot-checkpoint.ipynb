```julia
using DataFrames
using CSV
using Plots
using Images
```

## 1. Loading data

```julia
key = "tail"
```

```julia
sdm_folder_path = joinpath(@__DIR__, "Data", key, "SDM")
sdm_data_480_640 = CSV.read(joinpath(sdm_folder_path, "480_640", "sdm_data_zoom.csv"), DataFrame);
sdm_data_360_480 = CSV.read(joinpath(sdm_folder_path, "360_480", "sdm_data_zoom.csv"), DataFrame);
sdm_data_240_320 = CSV.read(joinpath(sdm_folder_path, "240_320", "sdm_data_zoom.csv"), DataFrame);
```

```julia
rse_folder_path = joinpath(@__DIR__, "Data", key, "RSE")
rse_data_480_640 = CSV.read(joinpath(rse_folder_path, "480_640", "rse_data_zoom.csv"), DataFrame);
rse_data_360_480 = CSV.read(joinpath(rse_folder_path, "360_480", "rse_data_zoom.csv"), DataFrame);
rse_data_240_320 = CSV.read(joinpath(rse_folder_path, "240_320", "rse_data_zoom.csv"), DataFrame);
```

```julia
S = sdm_data_480_640.S
# SSI
sdm_ssi = [sdm_data_480_640.ssi sdm_data_360_480.ssi sdm_data_240_320.ssi];
rse_ssi = [rse_data_480_640.ssi rse_data_360_480.ssi rse_data_240_320.ssi];
```

```julia
# SMPI
sdm_smpi = [sdm_data_480_640.smpi sdm_data_360_480.smpi sdm_data_240_320.smpi];
rse_smpi = [rse_data_480_640.smpi rse_data_360_480.smpi rse_data_240_320.smpi];
```

```julia
# ENL
sdm_enl = [sdm_data_480_640.enl sdm_data_360_480.enl sdm_data_240_320.enl];
rse_enl = [rse_data_480_640.enl rse_data_360_480.enl rse_data_240_320.enl];
```

## 2. Ploting
### 2.1 Ploting ENL

```julia
theme(:ggplot2)
pgfplotsx()
```

```julia
c = [RGB{Float64}(0.103874,0.056805,0.20243) RGB{Float64}(0.52005,0.369042,0.471546) RGB{Float64}(0.592425,0.683002,0.567492)]
```

```julia
plot(S, sdm_enl, 
    m = :pentagon, alpha = 0.9,
    xlabel = "S", ylabel = "ENL",
    legend = :none, color = c,
    size = (500, 450)
    )

p_ENL = plot!(S, rse_enl, 
    m = :c, alpha = 0.9, color = c,
    size = (500, 450)
    )

if key == "head"
    annotate!(180, sdm_enl[end, 1] + 0.015, Plots.text("480×640 (SDM)", 10, c[1]))
    annotate!(180, rse_enl[end, 1] - 0.017, Plots.text("480×640 (RSE)", 10, c[1]))

    annotate!(180, sdm_enl[end, 2] + 0.012, Plots.text("360×480 (SDM)", 10, c[2]))
    annotate!(180, rse_enl[end, 2] - 0.015, Plots.text("360×480 (RSE)", 10, c[2]))

    annotate!(180, sdm_enl[end, 3] + 0.012, Plots.text("240×320 (SDM)", 10, c[3]))
    annotate!(180, rse_enl[end, 3] - 0.015, Plots.text("240×320 (RSE)", 10, c[3]))
else
    annotate!(180, sdm_enl[end, 1] + 0.014, Plots.text("480×640 (SDM)", 10, c[1]))
    annotate!(180, rse_enl[end, 1] - 0.017, Plots.text("480×640 (RSE)", 10, c[1]))

    annotate!(180, sdm_enl[end, 2] + 0.015, Plots.text("360×480 (SDM)", 10, c[2]))
    annotate!(180, rse_enl[end, 2] - 0.015, Plots.text("360×480 (RSE)", 10, c[2]))

    annotate!(180, sdm_enl[end, 3] + 0.015, Plots.text("240×320 (SDM)", 10, c[3]))
    annotate!(180, rse_enl[end, 3] - 0.017, Plots.text("240×320 (RSE)", 10, c[3]))
end
```

### 2.2 Ploting SSI

```julia
plot(S, sdm_ssi, m = :pentagon, alpha = 0.9,
    xlabel = "S", ylabel = "SSI",
    legend = :none, color = c,
    size = (500, 450)
    )
p_SSI = plot!(S, rse_ssi, 
    m = :c, alpha = 0.9, color = c,
    size = (500, 450)
    )

if key == "head"
    annotate!(180, sdm_ssi[end, 1] - 0.006, Plots.text("480×640 (SDM)", 10, c[1]))
    annotate!(180, rse_ssi[end, 1] + 0.008, Plots.text("480×640 (RSE)", 10, c[1]))

    annotate!(180, sdm_ssi[end, 2] - 0.006, Plots.text("360×480 (SDM)", 10, c[2]))
    annotate!(180, rse_ssi[end, 2] + 0.006, Plots.text("360×480 (RSE)", 10, c[2]))

    annotate!(180, sdm_ssi[end, 3] - 0.006, Plots.text("240×320 (SDM)", 10, c[3]))
    annotate!(180, rse_ssi[end, 3] + 0.006, Plots.text("240×320 (RSE)", 10, c[3]))
else
    annotate!(180, sdm_ssi[end, 1] - 0.006, Plots.text("480×640 (SDM)", 10, c[1]))
    annotate!(180, rse_ssi[end, 1] + 0.008, Plots.text("480×640 (RSE)", 10, c[1]))

    annotate!(180, sdm_ssi[end, 2] - 0.007, Plots.text("360×480 (SDM)", 10, c[2]))
    annotate!(180, rse_ssi[end, 2] + 0.007, Plots.text("360×480 (RSE)", 10, c[2]))

    annotate!(180, sdm_ssi[end, 3] - 0.006, Plots.text("240×320 (SDM)", 10, c[3]))
    annotate!(180, rse_ssi[end, 3] + 0.008, Plots.text("240×320 (RSE)", 10, c[3]))
end
```

### 2.3 Ploting SMPI

```julia
#y_lims = Dict("head"-> (0.55, 0.86), "tail" -> ())
plot(S, sdm_smpi, m = :pentagon, alpha = 0.9,
    xlabel = "S", ylabel = "SMPI",
    legend = :none, color = c,
    )
p_SMPI = plot!(S, rse_smpi, 
    m = :c, alpha = 0.9, color = c,
    size = (500, 450),
    )

if key == "head"
    ylims!(0.55, 0.86)
    annotate!(180, sdm_smpi[end, 1] + 0.01, Plots.text("480×640 (SDM)", 10, c[1]))
    annotate!(180, rse_smpi[end, 1] + 0.012, Plots.text("480×640 (RSE)", 10, c[1]))

    annotate!(180, sdm_smpi[end, 2] + 0.01, Plots.text("360×480 (SDM)", 10, c[2]))
    annotate!(180, rse_smpi[end, 2] + 0.011, Plots.text("360×480 (RSE)", 10, c[2]))

    annotate!(180, sdm_smpi[end, 3] - 0.01, Plots.text("240×320 (SDM)", 10, c[3]))
    annotate!(180, rse_smpi[end, 3] - 0.01, Plots.text("240×320 (RSE)", 10, c[3]))
else
    annotate!(180, sdm_smpi[end, 1] - 0.008, Plots.text("480×640 (SDM)", 10, c[1]))
    annotate!(180, rse_smpi[end, 1] + 0.01, Plots.text("480×640 (RSE)", 10, c[1]))

    annotate!(180, sdm_smpi[end, 2] - 0.008, Plots.text("360×480 (SDM)", 10, c[2]))
    annotate!(180, rse_smpi[end, 2] + 0.008, Plots.text("360×480 (RSE)", 10, c[2]))

    annotate!(180, sdm_smpi[end, 3] - 0.008, Plots.text("240×320 (SDM)", 10, c[3]))
    annotate!(180, rse_smpi[end, 3] + 0.008, Plots.text("240×320 (RSE)", 10, c[3]))
end
```

```julia
p = plot(p_ENL, p_SSI, p_SMPI, layout=(1, 3), size=(900, 300), tex_output_standalone = true)
```

# 论文第五章的曲线图


## 1. loading data

```julia
theme(:ggplot2)
pgfplotsx()
```

```julia
key = "head"
path = joinpath(@__DIR__, "Data", key)
sdm_data = CSV.read(joinpath(path, "SDM", "480_640", "sdm_data_zoom.csv"), DataFrame);
rse_data = CSV.read(joinpath(path, "RSE", "480_640", "rse_data_zoom.csv"), DataFrame);
ldr_data = CSV.read(joinpath(path, "LDR", "480_640", "ldr_data_zoom.csv"), DataFrame);
S = sdm_data.S
enl = [sdm_data.enl rse_data.enl ldr_data.enl]
smpi = [sdm_data.smpi rse_data.smpi ldr_data.smpi]
ssi = [sdm_data.ssi rse_data.ssi ldr_data.ssi]
```

```julia
p1_enl = plot(S, enl, m = [:rect :pentagon :circle], alpha = 0.9, ms = 3, 
    xlabel = "S", ylabel = "ENL",
    legend = false, color = :auto,
    )
p1_ssi = plot(S, ssi, m = [:rect :pentagon :circle], alpha = 0.9, ms = 3, 
    xlabel = "S", ylabel = "SSI",
    legend = false, color = :auto,
    )
p1_smpi = plot(S, smpi, m = [:rect :pentagon :circle], alpha = 0.9, ms = 3, 
    xlabel = "S", ylabel = "SMPI",
    label = ["SDM" "RSE" "LDR"], color = :auto,
    )
```

```julia
p1 = plot(p1_enl, p1_ssi, p1_smpi, layout=(1, 3), size=(950, 300), tex_output_standalone = true)
```

```julia
saving_path = joinpath(@__DIR__, "Plot")
savefig(p1, joinpath(saving_path, "Chp5_compare_curve_$key.tex"))
```
