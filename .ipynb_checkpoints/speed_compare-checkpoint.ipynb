```julia
function sdm_speed(holo::AbstractMatrix, P::FFTW.cFFTWPlan, scalefactor::Number, Dx::Int, Dy::Int, dx::Int, dy::Int, N::Int)
    tensor = make_sub_holo(holo, Dx, Dy, dx, dy)
    @inbounds Threads.@threads for ind in CartesianIndices(1: size(tensor, 3))
        k = ind.I[1]
        tensor[:, :, k] = reconst(tensor[:, :, k], P, scale ÷ N^2)
    end
    rse_img = SDM(tensor)
end
```

```julia
function rse_speed(holo::AbstractMatrix, P::FFTW.cFFTWPlan, scalefactor::Number, Dx::Int, Dy::Int, dx::Int, dy::Int, N::Int)
    tensor = make_sub_holo(holo, Dx, Dy, dx, dy)
    @inbounds Threads.@threads for ind in CartesianIndices(1: size(tensor, 3))
        k = ind.I[1]
        tensor[:, :, k] = reconst(tensor[:, :, k], P, scale ÷ N^2)
    end
    rse_img = RSE_parallel(tensor)
end
```

```julia
function ldr_speed(holo::AbstractMatrix, P::FFTW.cFFTWPlan, scalefactor::Number, Dx::Int, Dy::Int, dx::Int, dy::Int, N::Int)
    tensor = make_sub_holo(holo, Dx, Dy, dx, dy; dims_shrink=true)
    @inbounds Threads.@threads for ind in CartesianIndices(1: size(tensor, 3))
        k = ind.I[1]
        tensor[:, :, k] = reconst(tensor[:, :, k], P, scale ÷ N^2)
    end
    rse_img = LDR(tensor, N)
end
```

```julia
PATH = joinpath(@__DIR__, "hologram")
holo_name = "head.bmp"
holo = load_image(PATH, holo_name)
Nx, Ny = size(holo)
scale = 4000000
P = plan_fft(holo)
re_img = reconst(holo, P, scale; shift=true)
Dx, Dy = Nx ÷ 2, Ny ÷ 2
@btime rse_speed(holo, P, 4000000, Dx, Dy, 50, 100, 2)
```

```julia
@btime sdm_speed(holo, P, 4000000, Dx, Dy, 50, 100, 2)
```

```julia
sub_P = plan_fft(similar(holo, Dx, Dy))
@btime ldr_speed(holo, sub_P, 4000000, Dx, Dy, 50, 100, 2)
```

# Effect of Random Shuffling (Used in Chapter 5)

```julia
function ldr_test(holo::AbstractMatrix, P::FFTW.cFFTWPlan, scalefactor::Number, Dx::Int, Dy::Int, dx::Int, dy::Int, N::Int)
    tensor = make_sub_holo(holo, Dx, Dy, dx, dy; dims_shrink=true)
    @inbounds Threads.@threads for ind in CartesianIndices(1: size(tensor, 3))
        k = ind.I[1]
        tensor[:, :, k] = reconst(tensor[:, :, k], P, scale ÷ N^2; shift=true)
    end
    rse_img = LDR(tensor, N)
end
```

## 1. generating data

```julia
PATH = joinpath(@__DIR__, "hologram")
holo_name = "tail.bmp"
holo = load_image(PATH, holo_name)
Nx, Ny = size(holo)
scale = 4000000
P = plan_fft(holo)
re_img = reconst(holo, P, scale; shift=true)
Dx, Dy = Nx ÷ 2, Ny ÷ 2
sub_P = plan_fft(similar(holo, Dx, Dy))
ssi, enl, smpi = Float64[], Float64[], Float64[]
key = holo_name[1:4]
ori_img, ori_img_zoom, ori_img_local = geometry_operate(re_img, holo_name)
num = 100

for _ in 1: num
    data = ldr_test(holo, sub_P, 4000000, Dx, Dy, 50, 100, 2)
    ldr_img, ldr_img_zoom, ldr_img_local = geometry_operate(data, holo_name)
    push!(ssi, SSI(ori_img_zoom, ldr_img_zoom))
    push!(enl, ENL(ldr_img_zoom))
    push!(smpi, SMPI(ori_img_zoom, ldr_img_zoom))
end
```

## 2. Ploting

```julia
theme(:ggplot2)
pgfplotsx()
```

```julia
scatter(1:num, ssi, m = :circle, alpha = 0.9, ms = 3, 
    xlabel = "S", ylabel = "SSI",
    legend = false, color = :auto,
    )
p_ssi = plot!([findmax(ssi)[2] findmin(ssi)[2]], [maximum(ssi) minimum(ssi)], m = :circle, alpha = 0.9, ms = 3, 
    xlabel = "S", ylabel = "SSI",
    legend = false, color = :red,
    )
scatter(1:num, enl, m = :circle, alpha = 0.9, ms = 3, 
    xlabel = "S", ylabel = "ENL",
    legend = false, color = :auto,
    )
p_enl = plot!([findmax(enl)[2] findmin(enl)[2]], [maximum(enl) minimum(enl)], m = :circle, alpha = 0.9, ms = 3, 
    xlabel = "S", ylabel = "ENL",
    legend = false, color = :red,
    )
scatter(1:num, smpi, m = :circle, alpha = 0.9, ms = 3, 
    xlabel = "S", ylabel = "SMPI",
    legend = false, color = :auto,
    )
p_smpi = plot!([findmax(smpi)[2] findmin(smpi)[2]], [maximum(smpi) minimum(smpi)], m = :circle, alpha = 0.9, ms = 3, 
    xlabel = "S", ylabel = "SMPI",
    legend = false, color = :red,
    )
p1 = plot(p_enl, p_ssi, p_smpi, layout=(3, 1), size = (500, 500), tex_output_standalone = true)
```

```julia
saving_path = joinpath(@__DIR__, "Plot")
savefig(p1, joinpath(saving_path, "scatter_$key.tex"))
```

## 3. export lo and hi data to CSV file

```julia
df = DataFrame(Method = ["ENL", "SSI", "SMPI"],
    Max = [maximum(enl), maximum(ssi), maximum(smpi)],
    Min = [minimum(enl), minimum(ssi), minimum(smpi)],
    error = [maximum(enl), maximum(ssi), maximum(smpi)] .- [minimum(enl), minimum(ssi), minimum(smpi)]
)
```

```julia
save(joinpath(saving_path, "hi_lo_$key.csv"), df)
```
