```julia
using DataFrames
using CSV
using Plots
using Images
```

## 1. Loading data

```julia
key = "tail"
```

```julia
sdm_folder_path = joinpath(@__DIR__, "Data", key, "SDM")
sdm_data_480_640 = CSV.read(joinpath(sdm_folder_path, "480_640", "sdm_data_zoom.csv"), DataFrame);
sdm_data_360_480 = CSV.read(joinpath(sdm_folder_path, "360_480", "sdm_data_zoom.csv"), DataFrame);
sdm_data_240_320 = CSV.read(joinpath(sdm_folder_path, "240_320", "sdm_data_zoom.csv"), DataFrame);
```

```julia
rse_folder_path = joinpath(@__DIR__, "Data", key, "RSE")
rse_data_480_640 = CSV.read(joinpath(rse_folder_path, "480_640", "rse_data_zoom.csv"), DataFrame);
rse_data_360_480 = CSV.read(joinpath(rse_folder_path, "360_480", "rse_data_zoom.csv"), DataFrame);
rse_data_240_320 = CSV.read(joinpath(rse_folder_path, "240_320", "rse_data_zoom.csv"), DataFrame);
```

```julia
S = sdm_data_480_640.S
sdm_ssi = [sdm_data_480_640.ssi sdm_data_360_480.ssi sdm_data_240_320.ssi];
rse_ssi = [rse_data_480_640.ssi rse_data_360_480.ssi rse_data_240_320.ssi];
```

```julia
# SMPI
sdm_smpi = [sdm_data_480_640.smpi sdm_data_360_480.smpi sdm_data_240_320.smpi];
rse_smpi = [rse_data_480_640.smpi rse_data_360_480.smpi rse_data_240_320.smpi];
```

```julia
# ENL
sdm_enl = [sdm_data_480_640.enl sdm_data_360_480.enl sdm_data_240_320.enl];
rse_enl = [rse_data_480_640.enl rse_data_360_480.enl rse_data_240_320.enl];
```

## 2. Ploting
### 2.1 Ploting ENL

```julia
theme(:ggplot2)
c = palette(:tokyo, 4)
```

```julia
plot(S, sdm_enl, 
    m=:circle, alpha=0.8, ms = 4.5, label = [" " "SDM" " "], 
    xlabel = "S", ylabel = "ENL",
    legend = :outertopleft, color = [c[1] c[2] c[3]]
    )

p_ENL = plot!(S, rse_enl, 
    m=:star5, ms = 4.5, alpha=0.8,
    label = [" " "RSE" " "], color = [c[1] c[2] c[3]]
    )

annotate!((204, sdm_enl[end, 1] + 0.015, Plots.text("480640", 10, c[1])))
annotate!((204, rse_enl[end, 1] - 0.015, Plots.text("480640", 10, c[1])))

annotate!((204, sdm_enl[end, 2] + 0.015, Plots.text("360480", 10, c[2])))
annotate!((204, rse_enl[end, 2] - 0.015, Plots.text("360480", 10, c[2])))

annotate!((204, sdm_enl[end, 3] + 0.015, Plots.text("240320", 10, c[3])))
annotate!((204, rse_enl[end, 3] - 0.015, Plots.text("240320", 10, c[3])))
```

### 2.2 Ploting SSI

```julia
plot(S, sdm_ssi, m=:circle, alpha=0.8, ms = 4.5, label = [" " "SDM" " "],
    xlabel = "S", ylabel = "SSI",
    legend = :outertopleft, color = [c[1] c[2] c[3]]
    )
p_SSI = plot!(S, rse_ssi, m=:diamond, ms = 4.5, alpha=0.8, 
    label = [" " "RSE" " "], color = [c[1] c[2] c[3]]
    )

annotate!((204, sdm_ssi[end, 1] - 0.006, Plots.text("480640", 10, c[1])))
annotate!((204, rse_ssi[end, 1] + 0.006, Plots.text("480640", 10, c[1])))

annotate!((204, sdm_ssi[end, 2] - 0.006, Plots.text("360480", 10, c[2])))
annotate!((204, rse_ssi[end, 2] + 0.006, Plots.text("360480", 10, c[2])))

annotate!((204, sdm_ssi[end, 3] - 0.006, Plots.text("240320", 10, c[3])))
annotate!((204, rse_ssi[end, 3] + 0.006, Plots.text("240320", 10, c[3])))
```

```julia
plot(S, sdm_smpi, m=:circle, alpha=0.8, ms = 4.5, label = [" " "SDM" " "], 
    xlabel = "S", ylabel = "SMPI",
    legend = :outertopleft, color = color = [c[1] c[2] c[3]]
    )
p_SMPI = plot!(S, rse_smpi, m=:diamond, ms = 4.5, alpha=0.8, 
    label = [" " "RSE" " "], color = [c[1] c[2] c[3]]
    )

annotate!((204, sdm_smpi[end, 1] - 0.008, Plots.text("480640", 10, c[1])))
annotate!((204, rse_smpi[end, 1] + 0.008, Plots.text("480640", 10, c[1])))

annotate!((204, sdm_smpi[end, 2] - 0.008, Plots.text("360480", 10, c[2])))
annotate!((204, rse_smpi[end, 2] + 0.008, Plots.text("360480", 10, c[2])))

annotate!((204, sdm_smpi[end, 3] - 0.008, Plots.text("240320", 10, c[3])))
annotate!((204, rse_smpi[end, 3] + 0.008, Plots.text("240320", 10, c[3])))
```
