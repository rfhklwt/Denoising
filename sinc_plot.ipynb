```julia
using Plots
theme(:ggplot2)
```

```julia
saving_path = joinpath(@__DIR__, "Plot")
```

#### Supported values for linetype: :contour, :heatmap, :mesh3d, :path, :path3d, :scatter, :scatter3d, :scattergl, :shape, :straightline, :surface, :wireframe

## 绘制sinc函数

```julia
x = range(-20, stop = 20, length = 3000)
y = range(-20, stop = 20, length = 3000)

f(x, y) = sinc(x)sinc(y)
```

```julia
sinc_plot = plot(x, y, f, linetype=:surface, camera = (60, 60))
```

```julia
sinc_plot_contour = plot(x, y, f, linetype=:contour, camera = (60, 60))
```

```julia
savefig(sinc_plot, joinpath(saving_path, "sinc_plot.pdf"))
```

```julia
savefig(sinc_plot_contour, joinpath(saving_path, "sinc_plot_contour.pdf"))
```
